```python
import kuramoto as ko
import matplotlib.pyplot as plt
import numpy as np
import pickle as pkl
```

# Fully connected

```python
dt = 0.01
tmax = 5

obj = ko.SpatialKuramoto(20, 10, "gaussian", w_mu=2, w_sgm=1, network="full")
obj.run(tmax=tmax, dt=dt)
r = ko.get_r(obj.phase_all)
```

```python
plt.figure(dpi=150, figsize=(5, 2))
plt.plot(obj.ts, np.sin(obj.phase_all[:, :10]), lw=1)
plt.xlabel("time", fontsize=12)
plt.ylabel(r"sin($\theta_{i}$)", fontsize=12)

plt.twinx()
plt.plot(obj.ts, r, 'k', lw=2)
plt.ylim([-0.1, 1.1])
plt.ylabel("R")

plt.show()
```

# Nearest neighbor

## weak interation

```python
dt = 0.01
tmax = 10

# 1, 20
obj = ko.SpatialKuramoto(10000, 1, "gaussian", w_mu=3, w_sgm=1, network="spatial")
obj.run(tmax=tmax, dt=dt)
```

```python
obj.save("./data/weak_example_K_1.pkl")
```

```python
video = ko.VideoMaker(obj)
video.export(out_name="weak_osc.mp4", frame_itv=5, fps=30)
```

```python
r = ko.get_r(obj.phase_all)

plt.figure(dpi=150, figsize=(6, 1))
plt.plot(obj.ts, r, 'k', lw=1)
plt.ylim([0, 0.05])
plt.xlabel("time (a.u.)", fontsize=14)
plt.ylabel("R", fontsize=14)
plt.show()
```
